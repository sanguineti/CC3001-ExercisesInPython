```python
%pylab inline
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,5))
fig.suptitle('Comparación entre Hanoi Tradicional y el Problema de Platos')

# Datos
n= linspace(1,8)
x1 = (2)*n
x2 = (2**(n/2))*(n - 3)
x3 = (2**n)
y =  (2**n) - 1

# Primer Gráfico, con s = 1
l1 = ax1.plot(n, x1, color="red", label='$2^s(n - s + 1)$ con $s=1$')[0]
l2 = ax1.plot(n, y, color="green", label='$2^s - 1$')
ax1.legend(loc='upper left')

# Segundo Gráfico, con s = n/2
l3 = ax2.plot(n, x2, color="blue",label='$2^s(n - s + 1)$ con $s= n/2$')[0]
l4 = ax2.plot(n, y, color="orange", label='$2^s - 1$')
ax2.legend(loc='upper left')

# Tercer Gráfico, con s = n
l5 = ax3.plot(n, x3, color="yellow", label='$2^s(n - s + 1)$ con $s=n$')[0]
l6 = ax3.plot(n, y, color="purple", label='$2^s - 1$')
ax3.legend(loc='upper left')

# Etiquetas de ejes
ax1.set_xlabel('Valor de N')
ax1.set_ylabel('Movimientos')
ax2.set_xlabel('Valor de N')
ax2.set_ylabel('Movimientos')
ax3.set_xlabel('Valor de N')
ax3.set_ylabel('Movimientos')

# Escala
plt.subplots_adjust(right=0.9)

# Mostrar el gráfico
plt.show()
fig.savefig('sample.png')
```
